1. Выведите количество пустых значений по колонкам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во пустых значений)

3. Выведите колонки, у которых медиана равна минимальному значению (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX.
Напишите какой вывод можно сделать по данным в этих колонках

```python
tasks+=1
pd.read_sql(
    '''
WITH new_table AS(
    SELECT * FROM (
        (SELECT 'CRIM' AS RES, AVG(CRIM) AS MED FROM (SELECT CRIM FROM boston ORDER BY CRIM LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(CRIM) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'ZN', AVG(ZN) AS MED FROM (SELECT ZN FROM boston ORDER BY ZN LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(ZN) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'INDUS', AVG(INDUS) AS MED FROM (SELECT INDUS FROM boston ORDER BY INDUS LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(INDUS) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'CHAS', AVG(CHAS) AS MED FROM (SELECT CHAS FROM boston ORDER BY CHAS LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(CHAS) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'NOX', AVG(NOX) AS MED FROM (SELECT NOX FROM boston ORDER BY NOX LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(NOX) AS MIN FROM boston))
    )
    
    SELECT * FROM new_table
    WHERE MED = MIN
    
    
    ''',
    con,
)
```

По результатам выборки, можно предположить что большая часть значений в столбцах CHAS и ZN имеют минимальное значение(0)

5. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг) среди всех значений LSTAT. Напишите какой вывод можно сделать по этим данным.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT * FROM
    (SELECT LSTAT, MEDV,  rank() OVER (ORDER BY LSTAT) AS RANG FROM boston)
    WHERE MEDV = (SELECT MAX(MEDV) FROM boston)
    ''',
    con,
)
```

По результатам выборки, можно сделать вывод, что более высокий статус, имеют владельцы дорогой недвижимости.

8. Выведите значения долей промышленной застройки(INDUS), концентрации оксидов азота(NOX) и по их перцентилям - 10, 20 ... 100 ( перцетиль(10,20...100),значение INDUS, значение NOX). Напишите прослеживается между ними взаимосвязь

```python
tasks+=1
pd.read_sql(
    '''
with select_1 AS (
        SELECT perc, MAX(INDUS) AS INDUS FROM (
            SELECT INDUS, (ntile(10) OVER (ORDER BY INDUS)) AS perc FROM boston
        ) GROUP BY perc
    ),
    select_2 AS(
        SELECT perc, MAX(NOX) AS NOX FROM (
            SELECT NOX, (ntile(10) OVER (ORDER BY NOX)) AS perc FROM boston
        ) GROUP BY perc
    )

SELECT select_2.perc * 10 AS PERC, INDUS, NOX FROM select_2 JOIN select_1 ON select_2.perc = select_1.perc


    ''',
    con,
)
```

По результатам выборки, можно сделать вывод, что чем больше строится индустриальной застройки, тем выше концентрация оксидов азота
