### –í—Å–µ –∑–∞–¥–∞–Ω–∏—è –Ω—É–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–æ–ª—å–∫–æ —Å –ø–æ–º–æ—â—å—é sql

1. –í—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º CRIM, ZN, INDUS, CHAS, NOX (–Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏, –∫–æ–ª-–≤–æ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π)

```python
tasks+=1
pd.read_sql(
    '''
    SELECT SUM(CRIM IS NULL) AS CRIM_NULL_ITOG,
           SUM(ZN IS NULL) AS ZN_NULL_ITOG,
           SUM(INDUS IS NULL) AS INDUS_NULL_ITOG,
           SUM(CHAS IS NULL) AS CHAS_NULL_ITOG,
           SUM(NOX IS NULL) AS NOX_NULL_ITOG
    FROM boston
    ''',
    con,
)
```

2. –í—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –∫–æ–ª–æ–Ω–æ–∫–∞–º CRIM, ZN, INDUS, CHAS, NOX (–Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏, –∫–æ–ª-–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π)

```python
tasks+=1
pd.read_sql(
    '''
    SELECT  COUNT(DISTINCT CRIM) as CRIM_unique_val,
            COUNT(DISTINCT ZN) as ZN_unique_val,
            COUNT(DISTINCT INDUS) as INDUS_unique_val,
            COUNT(DISTINCT CHAS) as CHAS_unique_val,
            COUNT(DISTINCT NOX) as NOX_unique_val
    FROM boston
    ''',
    con,
)
```

3. –í—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–æ–Ω–∫–∏, —É –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–¥–∏–∞–Ω–∞ —Ä–∞–≤–Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é (–Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏) –≤—ã–±–∏—Ä–∞—è –∏–∑ CRIM, ZN, INDUS, CHAS, NOX.
–ù–∞–ø–∏—à–∏—Ç–µ –∫–∞–∫–æ–π –≤—ã–≤–æ–¥ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –ø–æ –¥–∞–Ω–Ω—ã–º –≤ —ç—Ç–∏—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö

```python
tasks+=1
pd.read_sql(
    '''
WITH new_table AS(
    SELECT * FROM (
        (SELECT 'CRIM' AS RES, AVG(CRIM) AS MED FROM (SELECT CRIM FROM boston ORDER BY CRIM LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(CRIM) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'ZN', AVG(ZN) AS MED FROM (SELECT ZN FROM boston ORDER BY ZN LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(ZN) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'INDUS', AVG(INDUS) AS MED FROM (SELECT INDUS FROM boston ORDER BY INDUS LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(INDUS) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'CHAS', AVG(CHAS) AS MED FROM (SELECT CHAS FROM boston ORDER BY CHAS LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(CHAS) AS MIN FROM boston))
    UNION
    SELECT * FROM (
        (SELECT 'NOX', AVG(NOX) AS MED FROM (SELECT NOX FROM boston ORDER BY NOX LIMIT 2 OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))),
        (SELECT MIN(NOX) AS MIN FROM boston))
    )
    
    SELECT * FROM new_table
    WHERE MED = MIN
    
    
    ''',
    con,
)
```

–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –≤—ã–±–æ—Ä–∫–∏, –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å —á—Ç–æ –±–æ–ª—å—à–∞—è —á–∞—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö CHAS –∏ ZN –∏–º–µ—é—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ(0)

4. –í—ã–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–æ–º–Ω–∞—Ç(RM) –≤ –¥–æ–º–∞—Ö —Å —Å–∞–º–æ–π –¥–æ—Ä–æ–≥–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é(MEDV) –∏ 25 —Å–∞–º—ã–º–∏ –¥–µ—à–µ–≤—ã–º–∏ –¥–æ–º–∞–º–∏. 
–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –ø–æ 50, 100, 200, 300 —Å–∞–º—ã–º–∏ –¥–µ—à–µ–≤—ã–º–∏ –¥–æ–º–∞–º–∏. (–∫–æ–ª-–≤–æ –¥–æ–º–æ–≤(25,50,100,200,300), —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∫–æ–º–Ω–∞—Ç –≤ –Ω–∏—Ö, —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∫–æ–º–Ω–∞—Ç –≤ —Å–∞–º—ã—Ö –¥–æ—Ä–æ–≥–∏—Ö, —Ä–∞–∑–Ω–∏—Ü–∞).
–ù–∞–ø–∏—à–∏—Ç–µ –≤–ª–∏—è–µ—Ç –ª–∏ –∫–æ–ª-–≤–æ –∫–æ–º–Ω–∞—Ç –Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –∏ –∫–∞–∫ —Å–∏–ª—å–Ω–æ.

```python
tasks+=1
pd.read_sql(
    '''
SELECT (SELECT "25") as quantity_house,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston)) AS rich,
       (SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 25)) AS needy,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston))-(SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 25)) as margin
UNION ALL
SELECT (SELECT "50") as quantity_house,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston)) AS rich,
       (SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 50)) AS needy,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston))-(SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 50)) as margin
UNION ALL       
SELECT (SELECT "100") as quantity_house,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston)) AS rich,
       (SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 100)) AS needy,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston))-(SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 100)) as margin
UNION ALL       
SELECT (SELECT "200") as quantity_house,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston)) AS rich,
       (SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 200)) AS needy,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston))-(SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 200)) as margin
UNION ALL       
SELECT (SELECT "300") as quantity_house,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston)) AS rich,
       (SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 300)) AS needy,
       (SELECT avg(RM) FROM boston where MEDV = (SELECT max(MEDV)from boston))-(SELECT avg(RM) FROM (SELECT*from boston order by MEDV  limit 300)) as margin

''',
    con,
)
```

–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –≤—ã–±–æ—Ä–∫–∏, –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å —á—Ç–æ –∫–æ–ª-–≤–æ –∫–æ–º–Ω–∞—Ç –≤ –¥–æ–º–µ –≤–ª–∏—è–µ—Ç –Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–º–∞ –≤ —Ü–µ–ª–æ–º.–£ –±–æ–≥–∞—Ç—ã—Ö –∫–æ–º–Ω–∞—Ç –±–æ–ª—å—à–µ, —á–µ–º —É –Ω–∞—Å–µ–ª–µ–Ω–∏—è —Å –±–æ–ª–µ–µ –Ω–∏–∑–∫–∏–º —Å—Ç–∞—Ç—É—Å–æ–º(–±–µ–¥–Ω—ã—Ö).

5. –í—ã–≤–µ–¥–∏—Ç–µ —Ä–∞–Ω–≥–∏ –∑–Ω–∞—á–µ–Ω–∏–π –∫–æ–ª–æ–Ω–∫–∏ LSTAT(–ø—Ä–æ—Ü–µ–Ω—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è —Å –±–æ–ª–µ–µ –Ω–∏–∑–∫–∏–º —Å—Ç–∞—Ç—É—Å–æ–º) –≤ –¥–æ–º–∞—Ö —Å —Å–∞–º–æ–π –¥–æ—Ä–æ–≥–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é (–∑–Ω–∞—á–µ–Ω–∏–µ LSTAT, —Å—Ç–æ–∏–º–æ—Å—Ç—å, —Ä–∞–Ω–≥) —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π LSTAT. –ù–∞–ø–∏—à–∏—Ç–µ –∫–∞–∫–æ–π –≤—ã–≤–æ–¥ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –ø–æ —ç—Ç–∏–º –¥–∞–Ω–Ω—ã–º.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT * FROM
    (SELECT LSTAT, MEDV,  rank() OVER (ORDER BY LSTAT) AS RANG FROM boston)
    WHERE MEDV = (SELECT MAX(MEDV) FROM boston)
    ''',
    con,
)
```

–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –≤—ã–±–æ—Ä–∫–∏, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–π —Å—Ç–∞—Ç—É—Å, –∏–º–µ—é—Ç –≤–ª–∞–¥–µ–ª—å—Ü—ã –¥–æ—Ä–æ–≥–æ–π –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏.

6. –í—ã–≤–µ–¥–∏—Ç–µ —Å—Ä–µ–¥–Ω—é—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–º–æ–≤ –≥—Ä–∞–Ω–∏—á–∞—â–∏—Ö —Å —Ä–µ–∫–æ–π(CHAS) –∏ –Ω–µ—Ç (–≥—Ä–∞–Ω–∏—á–∏—Ç/–Ω–µ –≥—Ä–∞–Ω–∏—á–∏—Ç, —Å—Ç–æ–∏–º–æ—Å—Ç—å)

```python
tasks+=1
pd.read_sql(
    '''
    SELECT ROUND(AVG(MEDV),2) as PRICE_HOUSE, CHAS
    FROM boston
    GROUP BY CHAS 
    ''',
    con,
)
```

7. –í—ã–≤–µ–¥–∏—Ç–µ –≤—Å–µ –∫–æ–ª–æ–Ω–∫–∏, —É –∫–æ—Ç–æ—Ä—ã—Ö —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—à–µ, –∫–æ–≥–¥–∞ –¥–æ–º –≥—Ä–∞–Ω–∏—á–∏—Ç —Å —Ä–µ–∫–æ–π (–Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏) –≤—ã–±–∏—Ä–∞—è –∏–∑ CRIM, ZN, INDUS, CHAS, NOX. –ù–∞–ø–∏—à–∏—Ç–µ –∫–∞–∫–æ–π –≤—ã–≤–æ–¥ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –ø–æ —ç—Ç–∏–º –¥–∞–Ω–Ω—ã–º.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT AVG(CRIM), AVG(ZN), AVG(INDUS), AVG(CHAS), AVG(NOX) FROM boston
    ''',
    con,
)
```

```python
tasks+=1
pd.read_sql(
    '''
    SELECT AVG(CRIM), AVG(ZN), AVG(INDUS), AVG(CHAS), AVG(NOX) FROM boston
    WHERE CHAS = 1
    ''',
    con,
)
```

```python
tasks+=1
pd.read_sql(
    '''
with table_1 AS(SELECT * FROM 
    (SELECT 'CRIM' AS ITOG, ((SELECT AVG(CRIM) FROM boston WHERE CHAS = 1) > AVG(CRIM)) AS RES
    FROM boston
    UNION ALL
    SELECT 'ZN' AS ITOG, ((SELECT AVG(ZN) FROM boston WHERE CHAS = 1) > AVG(ZN)) AS RES
    FROM boston
    UNION ALL
    SELECT 'INDUS' AS ITOG, ((SELECT AVG(INDUS) FROM boston WHERE CHAS = 1) > AVG(INDUS)) AS RES
    FROM boston
    UNION ALL
    SELECT 'CHAS' AS ITOG, ((SELECT AVG(CHAS) FROM boston WHERE CHAS = 1) > AVG(CHAS)) AS RES
    FROM boston
    UNION ALL
    SELECT 'NOX' AS ITOG, ((SELECT AVG(NOX) FROM boston WHERE CHAS = 1) > AVG(NOX)) AS RES
    FROM boston
    ))
    SELECT * FROM table_1
    WHERE RES = 1
    ''',
    con,
)
```

–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –≤—ã–±–æ—Ä–∫–∏, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –ø—Ä–æ–º—ã—à–ª–µ–Ω–Ω—ã–µ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è —Å—Ç—Ä–æ—è—Ç—Å—è —É —Ä–µ–∫–∏ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–µ–Ω–Ω–æ –ø–æ–≤—ã—à–µ–Ω–∞ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –æ–∫—Å–∏–¥–æ–≤ –∞–∑–æ—Ç–∞.

8. –í—ã–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–ª–µ–π –ø—Ä–æ–º—ã—à–ª–µ–Ω–Ω–æ–π –∑–∞—Å—Ç—Ä–æ–π–∫–∏(INDUS), –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –æ–∫—Å–∏–¥–æ–≤ –∞–∑–æ—Ç–∞(NOX) –∏ –ø–æ –∏—Ö –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º - 10, 20 ... 100 ( –ø–µ—Ä—Ü–µ—Ç–∏–ª—å(10,20...100),–∑–Ω–∞—á–µ–Ω–∏–µ INDUS, –∑–Ω–∞—á–µ–Ω–∏–µ NOX). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ—Å–ª–µ–∂–∏–≤–∞–µ—Ç—Å—è –º–µ–∂–¥—É –Ω–∏–º–∏ –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å

```python
tasks+=1
pd.read_sql(
    '''
with select_1 AS (
        SELECT perc, MAX(INDUS) AS INDUS FROM (
            SELECT INDUS, (ntile(10) OVER (ORDER BY INDUS)) AS perc FROM boston
        ) GROUP BY perc
    ),
    select_2 AS(
        SELECT perc, MAX(NOX) AS NOX FROM (
            SELECT NOX, (ntile(10) OVER (ORDER BY NOX)) AS perc FROM boston
        ) GROUP BY perc
    )

SELECT select_2.perc * 10 AS PERC, INDUS, NOX FROM select_2 JOIN select_1 ON select_2.perc = select_1.perc


    ''',
    con,
)
```
